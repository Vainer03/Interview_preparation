### 8.1 Исследование модулей и объектов Python
В командной строке ввести python

````python
import datetime
print(dir(datetime))

# вывод атрибутов класса
print(dir(datetime.date))

# вывод конкретных атрибутов
print([_ for _ in dir(datetime) if 'date' in _.lower()])

# выводит документацию по модулю
print(help(datetime))

help(datetime.date)
help(datetime.date.fromtimestamp)
help(dir)
````

### 8.2 Изоляция зависимостей проекта при помощи virtualenv
Когда вы устанавливаете пакеты глобально, может существовать только одна версия библиотеки для всех ваших программ.

Разным программам могут понадобиться различные версии пакетов и даже самого python

Помимо этого, глобальная установка пакетов Python также может стать фактором рискка с точки зрения обеспечения безопасности. Для модификации глобальной среды нередко требуется выполнение команды pip install с правами суперпользователя. Когда вы устанавливаете новый пакет, то pip скачивает и исполняет код из интернета

#### Виртуальные среды спешат на помощь
Виртуальная среда - это изолированная среда Python. Физически располагается внутри папки, содержащей все пакеты и другие программные средства, от которых они зависят, в виде библиотек с нативным кодом и средой выполнения интерпретатора, в которых нуждаетя проект.

```
$ which pip3
/usr/local/bin/pip3
```

Создадим новую виртуальную среду
```
$ python3 -m venv ./venv
```
Посмотрим на содержимое папки
```
$ ls venv/
bin Include Lib pyvenv.cfg
```
Активируем виртуальную среду
```
$ source ./venv/bin/activate
(venv) $
```
Проверяем, что pip запускается из виртуальной среды
```
(venv) $ which pip3
/Users/dan/my-project/venv/bin/pip3
```
Питон тоже находится в виртуальной среде
```
(venv) $ which python
/Users/dan/my-project/venv/bin/python
```
Посмотрим на список установленных пакетов
```
(venv) $ pip list
pip (0.0.1)
setuptools (28.8.0)
```

```
(venv) $ pip install schedule
```
Как покинуть виртуальную среду
```
(venv) $ deactivate
$ which pip3
/usr/local/bin
```

### 8.3 По ту сторону байткода
Когда интерпретатор CPython исполняет вашу программу, он сначала ее транслирует в последовательность байткодовых инструкций. Байткод - это промежуточный язык для виртуальной машины Python, который используется в качестве оптимизации производительности.

````python
def greet(name):
    return 'Привет, ' + name + '!'

print(greet('Гвидо'))

print(greet.__code__.co_code)
print(greet.__code__.co_consts)
print(greet.__code__.co_varnames)

# disassembler
import dis
print(dis.dis(greet))
````
